```python
import pandas as pd
import numpy as np
import vectorbt as vbt
import plotly.graph_objects as go
from datetime import datetime
import os

def load_data(filepath='BTC_USDT_USDT.parquet'):
    """
    Загружает данные из Parquet. Если файла нет, генерирует демо-данные.
    Ожидает колонки: open_time, open, high, low, close, volume
    """
    if os.path.exists(filepath):
        print(f"✅ Загрузка данных из {filepath}...")
        df = pd.read_parquet(filepath)
        # Стандартизация имен колонок
        df.columns = df.columns.str.lower()
        if 'open_time' in df.columns:
            df['datetime'] = pd.to_datetime(df['open_time'], unit='ms')
        elif 'time' in df.columns:
            df['datetime'] = pd.to_datetime(df['time'], unit='ms')
        else:
            df['datetime'] = pd.to_datetime(df.index)
            
        df.set_index('datetime', inplace=True)
        df.sort_index(inplace=True)
    else:
        print(f"⚠️ Файл {filepath} не найден. Генерация тестовых данных...")
        # Генерация случайных данных для демонстрации (1 месяц 5м таймфрейм)
        dates = pd.date_range(start='2023-01-01', periods=8000, freq='5T')
        np.random.seed(42)
        close = 20000 + np.cumsum(np.random.randn(8000) * 10)
        high = close + np.random.rand(8000) * 20
        low = close - np.random.rand(8000) * 20
        open_p = close.shift(1).fillna(close)
        volume = np.random.randint(100, 1000, 8000)
        
        df = pd.DataFrame({
            'open': open_p, 'high': high, 'low': low, 'close': close, 'volume': volume
        }, index=dates)
        df.index.name = 'datetime'

    # Убедимся, что индекс datetime
    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index)
        
    return df

# ==========================================
# 2. РАСЧЕТ ИНДИКАТОРОВ (VECTORBT)
# ==========================================

def calculate_indicators(df):
    """
    Рассчитывает EMA 200 и RSI 14 с помощью vectorbt.
    """
    close = vbt.GenericData(df['close'].values)
    
    # EMA 200
    ema = vbt.MA.run(close, window=200, ma_type='ema')
    df['ema_200'] = ema.ma.values
    
    # RSI 14
    rsi = vbt.RSI.run(close, window=14)
    df['rsi_14'] = rsi.rsi.values
    
    return df

# ==========================================
# 3. ДВИЖОК ГЕНЕРАЦИИ ОРДЕРОВ (CUSTOM LOGIC)
# ==========================================

def generate_orders(df, commission=0.0005, max_trades_per_day=3):
    """
    Кастомная логика для соблюдения лимита сделок в день 
    и динамического SL/TP на основе High/Low свечи.
    """
    orders = []  # Список ордеров: {'time': ..., 'side': 1/-1, 'price': ..., 'size': ...}
    
    # Параметры стратегии
    sl_lookback = 5  # Смотрим назад на 5 свечей для SL
    rr_ratio = 2.0   # Risk Reward 1:2
    
    in_position = False
    position_side = 0  # 1 = Long, -1 = Short
    entry_price = 0.0
    sl_price = 0.0
    tp_price = 0.0
    trades_today = 0
    current_date = None
    
    # Предварительный расчет уровней SL для каждой свечи (для скорости)
    # Long SL: Min Low за последние N свечей
    # Short SL: Max High за последние N свечей
    df['min_low_5'] = df['low'].rolling(window=sl_lookback).min()
    df['max_high_5'] = df['high'].rolling(window=sl_lookback).max()
    
    data = df.to_dict('records')
    indices = df.index
    
    print("🔄 Симуляция торговли (по свечам)...")
    
    for i in range(sl_lookback, len(data) - 1):  # -1 чтобы смотреть вперед на выход
        row = data[i]
        next_row = data[i+1]
        current_time = indices[i]
        date_only = current_time.date()
        
        # Сброс лимита сделок при новом дне
        if current_date != date_only:
            current_date = date_only
            trades_today = 0
        
        # Если мы в позиции, проверяем выход (SL/TP)
        if in_position:
            # Проверяем свечу входа (i) и последующие, пока не выйдем
            # Для упрощения в этом цикле: мы проверяем выход на текущей свече (i), 
            # если вход был ранее. Но так как мы итерируемся, логичнее проверить 
            # условия выхода на High/Low текущей свечи относительно уровней.
            
            # В этой реализации мы открываем сделку на Open следующей свечи (i+1)
            # Поэтому проверку выхода делаем, когда позиция уже открыта.
            # Чтобы не усложнять цикл, мы будем открывать и закрывать в рамках одного прохода,
            # эмулируя "поиск выхода" вперед.
            pass 
            # Логика выхода перенесена в момент входа для точности бэктеста (Lookahead bias free внутри функции)
            # Ниже реализована логика "Вход -> Поиск выхода -> Запись ордеров"

        # Если мы НЕ в позиции, ищем вход
        if not in_position and trades_today < max_trades_per_day:
            # Условия входа
            trend_long = row['close'] > row['ema_200']
            trend_short = row['close'] < row['ema_200']
            
            rsi_oversold = row['rsi_14'] < 30
            rsi_overbought = row['rsi_14'] > 70
            
            # Паттерн: Зеленая свеча (Close > Open) для Long, Красная для Short
            bullish_candle = row['close'] > row['open']
            bearish_candle = row['close'] < row['open']
            
            signal_long = trend_long and rsi_oversold and bullish_candle
            signal_short = trend_short and rsi_overbought and bearish_candle
            
            if signal_long or signal_short:
                # Определение цены входа (Open следующей свечи)
                entry_price = next_row['open']
                direction = 1 if signal_long else -1
                
                # Расчет SL и TP
                if direction == 1:
                    # Long: SL по минимуму последних 5 свечей
                    sl_price = row['min_low_5']
                    # Защита от слишком широкого SL
                    if sl_price >= entry_price:
                        sl_price = entry_price * 0.995 
                else:
                    # Short: SL по максимуму последних 5 свечей
                    sl_price = row['max_high_5']
                    if sl_price <= entry_price:
                        sl_price = entry_price * 1.005
                
                # Расчет TP (RR 1:2)
                risk = abs(entry_price - sl_price)
                tp_price = entry_price + (risk * rr_ratio * direction)
                
                # Поиск точки выхода (Симуляция внутри свечей)
                exit_price = None
                exit_time = None
                
                # Проходим вперед, пока не найдем выход или не закончится день
                for j in range(i + 1, len(data)):
                    check_row = data[j]
                    check_high = check_row['high']
                    check_low = check_row['low']
                    check_time = indices[j]
                    
                    # Проверка SL/TP
                    hit_sl = False
                    hit_tp = False
                    
                    if direction == 1: # Long
                        if check_low <= sl_price:
                            exit_price = sl_price
                            hit_sl = True
                        elif check_high >= tp_price:
                            exit_price = tp_price
                            hit_tp = True
                    else: # Short
                        if check_high >= sl_price:
                            exit_price = sl_price
                            hit_sl = True
                        elif check_low <= tp_price:
                            exit_price = tp_price
                            hit_tp = True
                            
                    if hit_sl or hit_tp:
                        exit_time = check_time
                        break
                    
                    # Принудительное закрытие в конце дня, если не задето
                    if check_time.date() != current_date:
                        exit_price = check_row['close']
                        exit_time = check_time
                        break
                
                if exit_price is None: # Если цикл закончился (конец данных)
                    exit_price = data[-1]['close']
                    exit_time = indices[-1]
                
                # Запись ордеров
                # 1. BUY/SELL OPEN
                orders.append({
                    'time': next_row['open_time'] if 'open_time' in next_row else exit_time, # Используем время следующей свечи
                    'side': direction,
                    'price': entry_price,
                    'size': 1.0 # Нормализованный размер (1 BTC или 1 юнит)
                })
                
                # 2. SELL/BUY CLOSE
                orders.append({
                    'time': exit_time,
                    'side': -direction,
                    'price': exit_price,
                    'size': 1.0
                })
                
                trades_today += 1
                # В этой упрощенной модели мы не держим позицию между итерациями цикла,
                # так как сразу просчитали выход. 
                # in_position остается False, цикл продолжается.

    # Преобразование в DataFrame для vectorbt
    if not orders:
        return pd.DataFrame()
        
    orders_df = pd.DataFrame(orders)
    orders_df.set_index('time', inplace=True)
    orders_df.sort_index(inplace=True)
    
    # Векторизация для vectorbt: создаем матрицу ордеров
    # vectorbt ожидает колонки 'Size' (положительно для покупки, отрицательно для продажи) и 'Price'
    vbt_orders = pd.DataFrame(index=df.index)
    vbt_orders['Size'] = 0.0
    vbt_orders['Price'] = 0.0
    
    for idx, row in orders_df.iterrows():
        if idx in vbt_orders.index:
            # Если несколько ордеров в одну минуту, суммируем (для скальпинга редко, но возможно)
            vbt_orders.loc[idx, 'Size'] += row['side'] * row['size']
            vbt_orders.loc[idx, 'Price'] = row['price'] # Берем цену последнего ордера
            
    return vbt_orders

# ==========================================
# 4. ЗАПУСК БЭКТЕСТА И МЕТРИКИ
# ==========================================

def run_backtest():
    # 1. Данные
    df = load_data('BTC_USDT_USDT.parquet')
    
    # 2. Индикаторы
    df = calculate_indicators(df)
    
    # 3. Генерация ордеров
    orders_df = generate_orders(df, commission=0.0005, max_trades_per_day=3)
    
    if orders_df.empty or orders_df['Size'].sum() == 0:
        print("❌ Сделки не были сгенерированы. Проверьте условия входа или данные.")
        return

    # 4. Портфель VectorBT
    # Используем from_orders. Price берем из orders_df, Close из df для оценки PnL
    # Внимание: from_orders требует, чтобы цены ор
```
